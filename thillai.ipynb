```python
#scatter plot
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv("Student_Performance.csv")

# Print the dataset
print(df)

# Scatter Plot
plt.scatter(df["Attendance"], df["Marks"])

# Title
plt.title("Attendance vs Marks")

# X-axis
plt.xlabel("Attendance (%)")

# Y-axis
plt.ylabel("Marks")

# Show graph
plt.show()
```

```python
#Line chart
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv("Monthly_Attendance.csv")

# Print Dataset
print(df)

# Draw Line Chart
plt.plot(df["Month"], df["Attendance"], marker="*")

# Title
plt.title("Department of CSE-Overall Monthly Attendance")

# X-axis Label
plt.xlabel("Month")

# Y-axis Label
plt.ylabel("Attendance (%)")

# Grid
plt.grid(True)

# Display Graph
plt.show()
```

```python
#bar chart
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv("Student_Performance.csv")

# Print Dataset
print(df)

#colors = ["red","blue","green","orange"]

# Draw Bar Chart
plt.bar(df["Student"], df["Marks"], color="blue")

# Title
plt.title("Student Marks")


# X-axis Label
plt.xlabel("Student Name")

# Y-axis Label
plt.ylabel("Marks")

# Show Grid
plt.grid(axis='y')

# Display Graph
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr"]

sales = [120, 150, 170, 180]

profit = [30, 40, 45, 50]

plt.bar(months, sales, label="Sales", color="green")

plt.bar(months, profit, bottom=sales, label="Profit")

plt.legend()

plt.title("Stacked Bar Chart")

plt.show()
```

```python
#histogram
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv("Student_Performance.csv")

# Print Dataset
print(df)

# Draw Histogram
plt.hist(df["Marks"], bins=5)

# Title
plt.title("Distribution of Student Marks")

# X-axis Label
plt.xlabel("Marks")

# Y-axis Label
plt.ylabel("Frequency")

# Grid
plt.grid(True)

# Display Graph
plt.show()
```

```python
#oie chart
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv("Student_Performance.csv")

# Print Dataset
print(df)

# Count students in each department
dept = df["Department"].value_counts()

print(dept)

# Draw Pie Chart
plt.pie(dept,
        labels=dept.index,
        autopct="%1.1f%%")

# Title
plt.title("Department-wise Student Distribution")

# Display Graph
plt.show()
```

```python
#box plot
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv("Student_Performance.csv")

# Print Dataset
print(df)

# Draw Box Plot
plt.boxplot(df["Marks"])

# Title
plt.title("Student Marks Distribution")

# Y-axis Label
plt.ylabel("Marks")

# Grid
#plt.grid(True)
# Show Grid
plt.grid(axis='both')

# Display Graph
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file
df = pd.read_csv("Student_Performance.csv")

marks = df["Marks"]

# Calculate statistics
minimum = marks.min()
q1 = marks.quantile(0.25)
median = marks.median()
q3 = marks.quantile(0.75)
maximum = marks.max()

plt.figure(figsize=(7,6))

plt.boxplot(
    marks,
    patch_artist=True,
    boxprops=dict(facecolor="blue"),
    medianprops=dict(color="red", linewidth=2)
)

plt.title("Student Marks Distribution")
plt.ylabel("Marks")
plt.grid(axis='both', linestyle='--')

# Add labels
plt.text(1.08, minimum, f"Minimum = {minimum}", color="blue")
plt.text(1.08, q1, f"Q1 = {q1:.0f} (25%)", color="green")
plt.text(1.08, median, f"Median = {median:.0f} (50%)", color="red")
plt.text(1.08, q3, f"Q3 = {q3:.0f} (75%)", color="green")
plt.text(1.08, maximum, f"Maximum = {maximum}", color="blue")

plt.show()
```
