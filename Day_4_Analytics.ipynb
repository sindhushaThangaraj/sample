Data Visualization and Statistics
1. Random Sampling

Concept: Select random rows from a dataset.

```python
import pandas as pd

df = pd.DataFrame({
    "Name":["A","B","C","D","E"],
    "Marks":[85,70,90,65,88]
})

print(df.sample(2))
```

2. Handling Duplicates

```python
import pandas as pd

df = pd.DataFrame({
    "Name":["A","A","B","C"],
    "Marks":[90,90,80,75]
})

print(df.drop_duplicates())
```

3. Query Methods

Concept: Filter data using query().

```python
import pandas as pd

df = pd.DataFrame({
    "Name":["A","B","C"],
    "Marks":[80,65,90]
})

print(df.query("Marks > 70"))

```

astype("category") converts a column into a categorical data type, reducing memory usage and improving performance for repeated values.

4. Categories

Concept: Convert text data into categorical data.

```python
import pandas as pd

df = pd.DataFrame({
    "Grade":["A","B","A","C"]
})
print(df.dtypes)
df["Grade"] = df["Grade"].astype("category")

print(df.dtypes)
```

5. Vectorization

Concept: Perform operations on entire columns without loops.

```python
import pandas as pd

df = pd.DataFrame({
    "Salary":[25000,30000,40000]
})
df["Bonus"] = df["Salary"] * 0.10
print(df)
```

Matplotlib
6. Line Plot

Concept: Draw a simple line graph.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,30]

plt.plot(x,y)
plt.title("Line Plot")
plt.show()
```

7. Bar Plot

Concept: Compare categorical data.

```python
import matplotlib.pyplot as plt

subjects = ["Python","Java","C++"]
students = [45,30,25]

plt.bar(subjects,students)
plt.show()
```

8. Pie Chart

Concept: Show percentage contribution.

```python
import matplotlib.pyplot as plt

marks = [40,35,25]
labels = ["Python","Java","C++"]

plt.pie(marks,labels=labels,autopct="%1.1f%%")
plt.show()
```

9. Subplots

Concept: Display multiple plots together.

```python
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
plt.plot([1,2,3],[2,4,6])

plt.subplot(1,2,2)
plt.bar(["A","B"],[5,8])

plt.subplot(1,1,1)
plt.pie(marks,labels=labels,autopct="%1.1f%%")

plt.show()
```

10. Legends

Concept: Differentiate multiple plots.

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3],[2,4,6],label="Sales")
plt.plot([1,2,3],[1,3,5],label="Profit")

plt.legend()
plt.show()
```

Seaborn
11. Seaborn Line Plot

Concept: Statistical plotting using Seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.lineplot(data=tips,x="size",y="total_bill")

plt.show()
```

12. Seaborn Bar Plot

Concept: Average values for categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.barplot(data=tips,x="day",y="total_bill")

plt.show()
```

13. Seaborn Styles

Concept: Change graph appearance.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("darkgrid")

plt.plot([1,2,3],[4,2,5])

plt.show()
```

SciPy
14. Mean

Concept: Calculate average.

```python
from scipy import stats

data = [10,20,30,40,50]

print(stats.tmean(data))
```

15. Median

Concept: Find the middle value.

```python
import numpy as np

data = [5,10,15,20,25]

print(np.median(data))
```

16. Mode

Concept: Find the most frequent value.

```python
from scipy import stats

data = [1,2,2,3,4]

print(stats.mode(data))
```

Sampling Techniques
17. Random Sampling using NumPy

Concept: Randomly select values.

```python
import numpy as np

data = np.array([10,20,30,40,50])

sample = np.random.choice(data,3)

print(sample)
```

18. Random Numbers

Concept: Generate random numbers.

Combined Visualization Example

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    "Department":["IT","HR","Sales","Finance"],
    "Employees":[45,20,35,25]
})

sns.set_style("whitegrid")

plt.bar(df["Department"],df["Employees"])

plt.title("Employees by Department")
plt.xlabel("Department")
plt.ylabel("Employees")
plt.show()
```
