### Python Program to Demonstrate Tuples and Their Functions

Tuples are ordered, immutable collections of items. They are similar to lists but cannot be changed after creation.

```python
# 1. Creating a tuple
my_tuple = (1, 2, 3, 'apple', 'banana', 2)
print("Original tuple:", my_tuple)
print("Type of my_tuple:", type(my_tuple))

# Tuples can also be created without parentheses, often called tuple packing
another_tuple = 10, 20, 30
print("Another tuple (packed):", another_tuple)

# A tuple with a single item requires a comma
single_item_tuple = (42,)
print("Single item tuple:", single_item_tuple)
print("Type of single_item_tuple:", type(single_item_tuple))

# Empty tuple
empty_tuple = ()
print("Empty tuple:", empty_tuple)
```

```python
# 2. Accessing elements in a tuple (indexing and slicing)
print("\nAccessing Elements:")
print("First element (index 0):", my_tuple[0])
print("Last element (negative indexing):", my_tuple[-1])
print("Elements from index 1 to 3 (exclusive):", my_tuple[1:4])
```

### Tuple Immutability

Unlike lists, you cannot change the elements of a tuple after it's created. Trying to do so will raise a `TypeError`.

```python
# Attempting to change an element will result in an error
# Uncomment the line below to see the TypeError
#my_tuple[0] = 100
# print(my_tuple)

# You can, however, reassign the tuple variable to a new tuple
my_tuple = (5, 6, 7)
#my_tuple[1]=3
print("Tuple after reassignment:", my_tuple)
```

### Common Tuple Functions (Methods)

Tuples have limited built-in methods due to their immutability.

```python
# Using count()
original_tuple = (1, 2, 3, 'apple', 'banana', 2, 1, 2)
print("\nOriginal tuple for methods:", original_tuple)
print("Count of '2':", original_tuple.count(2))
print("Count of 'apple':", original_tuple.count('apple'))
print("Count of 'grape' (not in tuple):", original_tuple.count('grape'))

# Using index()
# Returns the index of the first occurrence of the specified value
print("Index of 'apple':", original_tuple.index('apple'))
print("Index of the first '2':", original_tuple.index(2))

# If the value is not found, it raises a ValueError
try:
    original_tuple.index('orange')
except ValueError as e:
    print(f"Error when finding 'orange': {e}")

# Finding length of a tuple
print("Length of the tuple:", len(original_tuple))
```

```python
# Tuple unpacking
coordinates = (10, 20)
x, y = coordinates
print(f"\nUnpacked coordinates: x={x}, y={y}")

# Iterating through a tuple
print("\nIterating through a tuple:")
for item in original_tuple:
    print(item)
```

### Python Program to Demonstrate Lists and Their Functions

Lists are ordered, mutable collections of items. They are one of the most versatile data types in Python.

```python
# 1. Creating a list
my_list = [10, 20, 30, 'orange', 'grape', 20]
print("Original list:", my_list)
print("Type of my_list:", type(my_list))

# Creating an empty list
another_list = []
print("Empty list:", another_list)

# List from another iterable (e.g., a string)
list_from_string = list("python")
print("List from string:", list_from_string)
```

```python
# 2. Accessing elements in a list (indexing and slicing)
print("\nAccessing Elements:")
print("First element (index 0):", my_list[0])
print("Last element (negative indexing):", my_list[-1])
print("Elements from index 1 to 4 (exclusive):", my_list[1:4])
```

### List Mutability

Unlike tuples, lists are mutable, meaning you can change their elements after creation.

```python
# 3. Modifying elements in a list
print("\nModifying Elements:")
print("List before modification:", my_list)
my_list[0] = 5
print("List after changing first element:", my_list)

# Modifying a slice
my_list[1:3] = [25, 35]
print("List after changing a slice:", my_list)
```

### Common List Functions (Methods)

Lists offer a rich set of methods for manipulation.

```python
# 4. Adding elements:
print("\nAdding Elements:")
my_list.append('kiwi') # Adds to the end
print("After append('kiwi'):", my_list)

my_list.insert(1, 'banana') # Inserts at a specific index
print("After insert(1, 'banana'):", my_list)

my_list.extend([100, 200]) # Appends elements from another iterable
print("After extend([100, 200]):", my_list)
```

```python
# 5. Removing elements:
print("\nRemoving Elements:")
my_list.remove('orange') # Removes the first occurrence of a value
print("After remove('orange'):", my_list)

popped_item = my_list.pop() # Removes and returns the last item
print("After pop() (item removed: " + str(popped_item) + "):", my_list)

popped_at_index = my_list.pop(0) # Removes and returns item at a specific index
print("After pop(0) (item removed: " + str(popped_at_index) + "):", my_list)

# To remove all items
# my_list.clear()
# print("After clear():", my_list)
```

```python
# 6. Other useful methods:
print("\nOther Useful Methods:")

# Count occurrences of a value
print("Count of 20:", my_list.count(20))

# Find index of a value
print("Index of 'grape':", my_list.index('grape'))

# Reversing the list
my_list.reverse()
print("After reverse():", my_list)

# Sorting the list (in-place)
numbers = [3, 1, 4, 1, 5, 9, 2]
print("Numbers before sort():", numbers)
numbers.sort()
print("Numbers after sort():", numbers)

# Getting the length of the list
print("Length of my_list:", len(my_list))

# Iterating through a list
print("\nIterating through a list:")
for item in my_list:
    print(item)
```
